```python
def separate_datatypes(df):
    cat=list(df.select_dtypes(include='object').columns)
    num=list(df.select_dtypes(exclude='object').columns)
    print(tabulate({'categorical':cat,'numerical':num},headers='keys'))
separate_datatypes(df)
```

#  univariate analysis

# display gender wise percentage in dataset

```python
fig,ax=plt.subplots(1,2,figsize=(15,8))
ax[0].set_title('count of gender')
percentage=df['gender'].value_counts()
labels=list(df['gender'].value_counts().index)

sns.countplot(x=df['gender'],ax=ax[0])
plt.pie(percentage,labels=labels,autopct='%0.2f%%')

plt.show()
```

# Display count of group of students

```python
plt.title('count of race/ethnicity')
sns.countplot(x=df['race/ethnicity'])
plt.show()
```

# Display percentage of parent level education

```python
plt.pie(df["parental level of education"].value_counts(), labels=df["parental level of education"].unique(),autopct="%.2f%%")
plt.title("Parental Level of Education")
plt.show()
```

# Display count of test preparation course

```python
plt.title('count of test preparation course')
sns.countplot(x=df['test preparation course'])
plt.show()
```

# Bi-variate analysis

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['reading score'],y=df['writing score'],hue=df.gender)
plt.show()
```

```python
scores = [df['math score'],df['reading score'],df['writing score']]
plt.boxplot(scores, labels=['math', 'read', 'write'])
plt.xlabel('Score Type')
plt.ylabel('Total Score')
plt.title('Distribution of Total Scores')
plt.show()
```

```python
ax=sns.countplot(x=df['test preparation course'],hue=df['gender'],data=df)
plt.show()
```

```python
ax=sns.countplot(x=df['race/ethnicity'],hue=df['gender'],data=df)
plt.show()
```

# Multi-variate analysis

```python
print(df['math score'].max())
print(df['math score'].min())
```

```python
mv=df[(df['math score']>60)&(df['reading score']>40)]
mv
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,6))
plt.title('count of test preparation course')
count=mv['test preparation course'].value_counts()
print(count)
label=['not-completed','completed']
ax=sns.countplot(x=mv['test preparation course'],ax=ax[0])
plt.pie(count,labels=label,autopct='%0.2f%%')

plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,7))
sns.histplot(x=mv['math score'],hue=mv['gender'],ax=ax[0],bins=6)
sns.boxplot(y=mv['math score'],x=mv['gender'],ax=ax[1])
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,7))
sns.histplot(x=mv['math score'],hue=mv['lunch'],ax=ax[0],bins=6)
sns.boxplot(y=mv['math score'],x=mv['lunch'],ax=ax[1])
plt.show()
```
